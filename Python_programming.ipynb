```python
# Loading the libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Loading the data 
training_set = pd.read_csv('train.csv')
test_set = pd.read_csv('test.csv')
ideal_set = pd.read_csv('ideal.csv')
```

# Finding the four ideal functions 
"[The] task is to write a Python-program that uses training data to choose the four ideal functions which are the best fit out of the fifty provided."

```python
# Modifying the names of the columns to not confuse them with the ideal functions
training_set.columns = ['x', 'function1', 'function2', 'function3', 'function4']
training_set_h= training_set.transpose()
ideal_set_h = ideal_set.transpose()
```

```python
# Create a class for the training functions 
class Function:
  def __init__(self, dataframe, position):
    '''to do '''
    self.x_data = dataframe.iloc[0]
    self.y_data = dataframe.iloc[position]
  
  def show(): 
    pass 

# Create a class for the ideal functions 
class IdealFunction(Function):
  def __init__(self, dataframe, position):
    super().__init__(dataframe, position)
    self.full = dataframe
```

```python
# Create the four instances with our training datas
f1_training, f2_training, f3_training, f4_training = Function(training_set_h, 1), Function(training_set_h, 2), Function(training_set_h, 3), Function(training_set_h, 4)
```

```python
# Having a look at the 4 training functions
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15,10))
axes[0,0].plot(f1_training.x_data , f1_training.y_data)
axes[0, 0].set_title("Function 1")
axes[0,1].plot(f2_training.x_data, f2_training.y_data)
axes[0, 1].set_title("Function 2")
axes[1,0].plot(f3_training.x_data, f3_training.y_data)
axes[1, 0].set_title("Function 3")
axes[1,1].plot(f4_training.x_data, f4_training.y_data)
axes[1, 1].set_title("Function 4")
plt.show()
```

```python
# Create dataframes that include the function you want to analyse and all the ideal functions
y1ideal = pd.concat([ideal_set, f1_training.y_data], axis=1)
y2ideal = pd.concat([ideal_set, f2_training.y_data], axis=1)
y3ideal = pd.concat([ideal_set, f3_training.y_data], axis=1)
y4ideal = pd.concat([ideal_set, f4_training.y_data], axis=1)
```

```python
# Create 4 dataframes to store the results
resultsy1, resultsy2, resultsy3, resultsy4 = pd.DataFrame(), pd.DataFrame(), pd.DataFrame(), pd.DataFrame()
'''Goes into each dataframe and for each function calculates the least square, 
i.e. it takes the difference between the y-value of the ideal function and 
that of our function for every point and every function'''
for i in range(1, 51):
  name = 'y' + str(i) + 'square'
  resultsy1[name] = y1ideal.apply(lambda row: (row.iloc[i] - row.function1)**2, axis=1)
  resultsy2[name] = y2ideal.apply(lambda row: (row.iloc[i] - row.function2)**2, axis=1)
  resultsy3[name] = y3ideal.apply(lambda row: (row.iloc[i] - row.function3)**2, axis=1)
  resultsy4[name] = y4ideal.apply(lambda row: (row.iloc[i] - row.function4)**2, axis=1)
# Calculate the sum
sumy1 = resultsy1.sum()
sumy2 = resultsy2.sum()
sumy3 = resultsy3.sum()
sumy4 = resultsy4.sum()
```

```python
# Here is the list of each ideal functions ranked according to the sum of the squares errors for the function1
sumy1.sort_values()
```

```python
# The best fit according to the least square method is the function 47
ideal47 = IdealFunction(ideal_set_h, 47)
# Let's compare the two functions visually 
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,8))
axes[0].plot(ideal47.x_data, ideal47.y_data)
axes[0].set_title('Ideal function 47')
axes[1].plot(f1_training.x_data, f1_training.y_data)
axes[1].set_title('Training function 1')
fig.show()
```

# Finding the delta max for each function
'The criterion for mapping the individual test case to the four ideal functions is that the existing maximum deviation of the calculated regression does not exceed the largest deviation between training dataset (A) and the ideal function (C) chosen for it by more than factor sqrt(2)'.

```python
# Create a dataframe to store the delta max of each function
delta = pd.DataFrame()
delta['function 1'] = abs(ideal47.y_data - f1_training.y_data)
delta['function 2'] = abs(ideal48.y_data - f2_training.y_data)
delta['function 3'] = abs(ideal32.y_data - f3_training.y_data)
delta['function 4'] = abs(ideal40.y_data - f4_training.y_data)
deltamax1 = max(delta['function 1'])
deltamax2 = max(delta['function 2'])
deltamax3 = max(delta['function 3'])
deltamax4 = max(delta['function 4'])
```

```python
print("The delta max for function 1 is {}".format(deltamax1))
print("The delta max for function 2 is {}".format(deltamax2))
print("The delta max for function 3 is {}".format(deltamax3))
print("The delta max for function 4 is {}".format(deltamax4))
```

```python
# Let's visualize the four ideal functions with their bandwith
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20,8))
axes[0,0].plot(ideal47.x_data, ideal47.y_data)
axes[0,0].fill_between(ideal47.x_data, ideal47.y_data - deltamax1, ideal47.y_data + deltamax1, color= 'yellow', alpha= 0.5)
axes[0,1].plot(ideal48.x_data, ideal48.y_data)
axes[0,1].fill_between(ideal48.x_data, ideal48.y_data - deltamax2, ideal48.y_data + deltamax2, color= 'yellow', alpha= 0.5)
axes[1,0].plot(ideal32.x_data, ideal32.y_data)
axes[1,0].fill_between(ideal32.x_data, ideal32.y_data - deltamax3, ideal32.y_data + deltamax3, color= 'yellow', alpha= 0.5)
axes[1,1].plot(ideal40.x_data, ideal40.y_data)
axes[1,1].fill_between(ideal40.x_data, ideal40.y_data - deltamax4, ideal40.y_data + deltamax4, color= 'yellow', alpha= 0.5)
fig.show()
```

```python
ideal47x = pd.DataFrame()
ideal47x['x']= ideal47.x_data
ideal47x['y']= ideal47.y_data
ideal47x=ideal47x.set_index('x')
ideal48x = pd.DataFrame()
ideal48x['x']= ideal48.x_data
ideal48x['y']= ideal48.y_data
ideal48x=ideal48x.set_index('x')
ideal32x = pd.DataFrame()
ideal32x['x']= ideal32.x_data
ideal32x['y']= ideal32.y_data
ideal32x=ideal32x.set_index('x')
ideal40x = pd.DataFrame()
ideal40x['x']= ideal40.x_data
ideal40x['y']= ideal40.y_data
ideal40x=ideal40x.set_index('x')
```

```python
def find_coordinates(dataframe, deltamax) : 
  results = pd.DataFrame()
  for x,y in test_set.values:
    y_ideal = dataframe.loc[x][0]
    y_max = y_ideal+deltamax
    y_min = y_ideal-deltamax
    if y_min <= y <= y_max:
      coord = {'x': x, 'y': y}
      results = results.append(coord, ignore_index=True)
  return results
```

```python
points_ideal47 = find_coordinates(ideal47x, deltamax1)
points_ideal48 = find_coordinates(ideal48x, deltamax2)
points_ideal32 = find_coordinates(ideal32x, deltamax3)
points_ideal40 = find_coordinates(ideal40x, deltamax4)
```
